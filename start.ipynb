```python
import polars as pl

import json

from glob import glob
from tqdm.notebook import tqdm
```

```python
SPLIT='train'
SPLIT='val'
SPLIT='test'
```

```python
records = []
counter = 0
for datasetn in glob(f'/mnt/ssl/{SPLIT}/*'):
    dataset = datasetn.split('/')[-1]
    #datase: spring-pbmcs-inflammasomescreen
    celltype = dataset.split('-')[1]
    print(counter, dataset)
    for stainn in glob(datasetn + '/*'):
        stain = stainn.split('/')[-1]
        for platen in glob(stainn + '/*'):
            plate = platen.split('/')[-1]
            for filen in glob(platen + '/*'):
                file = filen.split('/')[-1]
                if not (file.endswith('.tiff') or file.endswith('.tif')): continue
                #file: A01_f02_577_227.tiff
                well, field, x, y = file.split('.')[0].split('_')
                #print(dataset, stain, plate, file, well, field, x, y, filen)
                records.append((counter, dataset, celltype, stain, plate, file, well, field, x, y, filen))
                counter +=1
                #break
            #break
        #break
    #break
```

```python
len(records)
```

```python

df_records = pl.DataFrame(records, schema=['id', 'dataset', 'celltype', 'stain', 'plate', 'file', 'well', 'field', 'x', 'y', 'path'])
```

```python
df_records.shape
```

```python
df_records['dataset'].value_counts()
```

```python
df_records.write_csv(f'/mnt/ssl/records_{SPLIT}.csv')
```

```python
id_to_path = df_records[['id', 'path']].to_dict()
#make a dictionary from a list of keys and values
id_to_path_ = dict(zip(id_to_path['id'], id_to_path['path']))
json.dump(id_to_path_, open(f'/mnt/ssl/{SPLIT}_id_to_path.json', 'w'))
```
